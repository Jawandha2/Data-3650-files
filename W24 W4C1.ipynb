## Interacting with Databases

```python
import sqlite3
```

```python
conn = sqlite3.connect('players.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute("""CREATE TABLE player(
name VARCHAR(20),
jersey INTEGER,
team VARCHAR(5)
)""")
```

```python
conn.commit()
```

```python
cur.execute("INSERT INTO player VALUES ('Connor', 97, 'EO')")
conn.commit()
```

```python
cur.execute('SELECT * FROM player')
```

```python
row = cur.fetchall()
row
```

```python
cur.execute("INSERT INTO player VALUES ('Leon', 29, 'EO')")
cur.execute("INSERT INTO player VALUES ('Nazem', 91, 'CF')")
conn.commit()
```

```python
cur.execute('SELECT * FROM player')
row = cur.fetchall()
row
```

```python
cur.description
```

```python
col = [i[0] for i in cur.description]
col
```

```python
df = pd.DataFrame(data = row, columns= col)
df
```

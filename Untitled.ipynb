## Extreme Values

```python
np.random.seed(101)
data = pd.DataFrame(np.random.randn(1000, 4), columns=['One', 'Two', 'Three', 'Four'])
data
```

```python
data.describe()
```

```python
data['Two'][(data['Two'] < -3) | (data['Two'] > 3)]
```

```python
data[np.abs(data['Two']) > 3]['Two']
```

```python
data
```

```python
# records for which any column values fall outside the range of -3 to +3
data[data[np.abs(data) > 3].any(1)]
```

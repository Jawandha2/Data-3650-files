# Data Visualization using pandas and seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Data Visualization using pandas

```python
np.random.seed(10)
ser = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0, 100, 10))
ser
```

```python
ser.plot()
```

```python
np.random.seed(1000)
ser = pd.Series(np.random.rand(16), index = 'a b c d e f g h i j k l m n o p'.split())
ser
```

```python
fig, axes = plt.subplots(2, 1)
ser.plot.bar(color = 'g', ax = axes[0])
ser.plot.barh(color = 'r', ax = axes[1])
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                 index = 'one two three four five six'.split(),
                 columns = pd.Index(['A', 'B', 'C', 'D']))
df
```

```python
df.plot.bar()
```

```python
df[['A', 'B']].plot.bar()
```

### Data Visualization using Seaborn

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

#### Showing data distribution: displot()

```python
sns.displot(df['total_bill'], bins = 30)
```

```python
sns.displot(df['total_bill'], kind = 'kde', fill = True)
```

```python
sns.displot(df['tip'])
```

```python
# facetgrid column
sns.displot(x = 'tip', data = df, col = 'sex')
```

```python
sns.displot(x = 'tip', data = df, col = 'sex', row = 'time')
```

```python
sns.displot(x = 'tip', data = df, col = 'sex', row = 'time', height = 3)
```

```python
sns.displot(data = df, x = 'tip', col = 'sex', row = 'time', height = 3, aspect = 1.5)
```

### jointplot() and pairplot()

```python
sns.jointplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.pairplot(data = df)
```

```python
df
```

#### countplot()

```python
sns.countplot(data = df, x = 'smoker')
```

```python
sns.boxplot(data = df, x = 'day', y = 'total_bill')
```

```python
sns.boxplot(data = df, x = 'day', y = 'total_bill', hue = 'sex')
```

### colors and styles
[color palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)

```python
sns.catplot(data = df, x = 'day', y = 'tip', hue = 'sex', palette = 'magma')
```

### relplot()
- displaying the relationship between two int variables
- similar to scatter plot()

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip',
           sizes = (100, 350), 
           size = 'size',
           color = 'orange',
           edgecolor = 'black',
           alpha = 0.7)
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip',
           hue = 'sex',
            style = 'smoker',
            alpha = 0.7,
           palette = ['royalblue', 'deeppink'])
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip',
           marker = '*',
           size = 'size',
           sizes = (100, 300))
```

```python
# lengeds - show the identity of a graph
data = np.arange(2,20,2)
data
```

```python
square = data ** 2
cube = data ** 3
print(square)
print(cube)
```

```python
plt.plot(data, square, marker = 'o', label = 'Square')
plt.plot(data, cube, marker = 'o', label = 'Cube')
plt.xlabel('input data')
plt.ylabel('output data')
plt.title('This is a title')
plt.legend()
plt.show()
```

```python
plt.plot(data, square, marker = 'o')
plt.plot(data, cube, marker = 'o')
plt.xlabel('input data')
plt.ylabel('output data')
plt.title('This is a title')
plt.legend(['Square', 'Cube'])
plt.show()
```

```python
line1, = plt.plot(data, square, marker = 'o')
line2, = plt.plot(data, cube, marker = 'o')
plt.xlabel('input data')
plt.ylabel('output data')
plt.title('This is a title')
plt.legend([line1, line2],['Square', 'Cube'], loc = 'upper left', ncol = 2, frameon = False)
plt.show()
```

## Subplots
- subplot controls the layout of a figure
- layout: # of rows/columns
- 3 parameters
- plt.subplot(1, 2, 1)

```python
# figure can have multiple axes
# when you do not provide any argument in a subplot(), this means you will have 1 row and 1 column
fig, axes = plt.subplots()

axes.plot(data, square, marker = 'o', label = 'Square')
axes.plot(data, cube, marker = 'o', label = 'Cube')


plt.xlabel('input data')
plt.ylabel('output data')
plt.title('This is a title')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(data, square, marker = 'o', label = 'Square')
axes[1].plot(data, cube, marker = 'o', label = 'Cube')

axes[0].set_xlabel('First X axis')
axes[0].set_ylabel('First Y axis')
axes[0].set_title('Fist title')

axes[1].set_xlabel('Second X axis')
axes[1].set_ylabel('Second Y axis')
axes[1].set_title('Second title')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

axes[0].plot(data, square, color='r', marker = 'o', label = 'Square')
axes[1].plot(data, cube, 'go-.', label = 'Cube')

axes[0].set_xlabel('First X axis')
axes[0].set_ylabel('First Y axis')
axes[0].set_title('First title')
axes[0].grid()

axes[1].set_xlabel('Second X axis')
axes[1].set_ylabel('Second Y axis')
axes[1].set_title('Second title')
axes[1].grid()

plt.suptitle('This is a common title')
plt.tight_layout()
plt.show()
```
